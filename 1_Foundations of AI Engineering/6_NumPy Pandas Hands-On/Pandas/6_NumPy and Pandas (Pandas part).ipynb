## fillna

fill not available

```python
new_df = df.fillna(0)

new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No Event'
})

new_df
```

```python
## foward fill


new_df = df.fillna(method='ffill')

new_df
```

```python
## drop na

new_df = df.dropna()

new_df
```

```python
new_df = df.dropna(how='all')

new_df
```

```python
new_df = df.dropna(thresh=2)

new_df
```
